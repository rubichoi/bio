```python
charge2 = open("C:/Users/ADMIN/Desktop/실습자료/FDR/201909r.1.real.charge2.txt",'r')
fdrResult = open("C:/Users/ADMIN/Desktop/실습자료/FDR/201909r.1.real.charge2.1%.txt",'w')
tcount = 0
dcount = 0
XXXcount = 0.0
bigValue = 0.0
tscount = 0.0
dscount = 0.0
checkCount = 0
targetList = []
sortResult = []
removeDuplicate = []

for line in charge2:
    if line[0] == "C" or line[0] =="s":
        fdrResult.write(line)
    else:
        sortResult.append(line.split("\t"))
        
sortResult = sorted(sortResult, key = lambda sortResult:int(sortResult[0]))

for i in range(0, len(sortResult)):
    if len(removeDuplicate) == 0:
        removeDuplicate.append(sortResult[i])
    else:
        if sortResult[i][0] == removeDuplicate[checkCount-1][0]:
            if float(sortResult[i][5]) < float(removeDuplicate[checkCount-1][5]):
                removeDuplicate[checkCount-1] =sortResult[i]
            checkCount = checkCount -1
            
        else:
            removeDuplicate.append(sortResult[i])
    checkCount = checkCount +1

#for i in range(0,len(removeDuplicate)):
#    fdrResult.write("\t".join(removeDuplicate[i]).rstrip("\t"))

sortResult = sorted(removeDuplicate, key = lambda removeDuplicate:float(removeDuplicate[5]))

#for i in range(0,len(sortResult)):
#    fdrResult.write("\t".join(sortResult[i]).rstrip("\t"))
    
for i in range(0, len(sortResult)):
    for j in range(0,len(sortResult[i][15].split(","))):
        if "XXX_" in sortResult[i][15].split(",")[j]:
            XXXcount = XXXcount + 1.0
            
    if XXXcount == len(sortResult[i][15].split(",")):
        dcount = dcount + 1.0
    else:
        targetList.append("\t".join(sortResult[i]).rstrip("\t"))
        tcount = tcount + 1.0
    

    XXXcount =0.0

    if tcount != 0.0 and dcount != 0.0:
        if dcount/tcount <= 0.01 and dcount/tcount >= bigValue:
            tscount = tcount
            dscount = dcount
            
for i in range(0,int(tscount)):
    fdrResult.write(targetList[i])

print("targetCount:" + str(tcount))
print("decoyCount:" + str(dcount))
print("e-value threshold = " + str(float(targetList[int(tscount)-1].split("\t")[5])))
print("total target : " + str(tscount))
        
charge2.close()
fdrResult.close()
```
